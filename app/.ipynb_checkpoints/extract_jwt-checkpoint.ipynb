```python
import subprocess
import tempfile
import time
import os
import json

# EXE 路徑
exe_path = r"E:\AP\Api2\api2.exe"

# 啟動 exe（非阻塞）
proc = subprocess.Popen(exe_path)

# 等待 exe 解壓暫存資料夾
time.sleep(2)  # 可依 exe 啟動速度調整秒數

# 找到 TEMP 下的 _MEIxxxxxx 資料夾
temp_dir = tempfile.gettempdir()
mei_folders = [f for f in os.listdir(temp_dir) if f.startswith("_MEI")]
jwt_key = None

for folder in mei_folders:
    mei_path = os.path.join(temp_dir, folder)
    security_json = os.path.join(mei_path, "security.json")
    if os.path.exists(security_json):
        with open(security_json, "r") as f:
            data = json.load(f)
            jwt_key = data.get("Jwt", {}).get("Key")
            print(f"Found JWT Key: {jwt_key}")
        break

if not jwt_key:
    print("JWT Key not found.")

# 可選：關閉 exe
proc.terminate()
```
