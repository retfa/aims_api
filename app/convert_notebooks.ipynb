```python
# convert_notebooks_api.py
from nbconvert import PythonExporter
import nbformat
import os
import time

NOTEBOOK_DIR = r"C:\Users\Jason.Ouyang\Downloads\OuYang\Python\20250204_PM21_GreenZone\app"
NOTEBOOK_DIR_LLM = r"C:\Users\Jason.Ouyang\Downloads\OuYang\Python\oldWorkData\LLM"

def convert_notebooks_api(base_dir):
    for root, dirs, files in os.walk(base_dir):
        # 忽略 .ipynb_checkpoints 資料夾
        dirs[:] = [d for d in dirs if d != ".ipynb_checkpoints"]

        for file in files:
            if file.endswith(".ipynb") and "convert_notebooks" not in file:
                notebook_path = os.path.join(root, file)
                py_path = notebook_path.replace(".ipynb", ".py")

                # 只轉最近修改的 notebook
                if os.path.exists(py_path):
                    if os.path.getmtime(py_path) >= os.path.getmtime(notebook_path):
                        # .py 比 notebook 新 → 跳過
#                         print(f"略過（未修改）: {notebook_path}")
                        continue

                print(f"正在轉換: {notebook_path}")

                # 讀 notebook
                nb_node = nbformat.read(notebook_path, as_version=4)
                exporter = PythonExporter()
                source, _ = exporter.from_notebook_node(nb_node)

                # 寫入 .py
                with open(py_path, "w", encoding="utf-8") as f:
                    f.write(source)

    print("全部轉換完成！")

if __name__ == "__main__":
    start_time = time.time()
    
    print('=============================開始轉換FASTAPI==========================')

    convert_notebooks_api(NOTEBOOK_DIR)
    
    print('=============================開始轉換LLM==========================')
    
    convert_notebooks_api(NOTEBOOK_DIR_LLM)

    end_time = time.time() - start_time
    print("總轉換時間:", end_time)
```
